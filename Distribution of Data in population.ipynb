# Visualization

```python
import seaborn as sns
import matplotlib 
import matplotlib.pyplot as plt
```

```python
sns.set_style('darkgrid')
```

```python
# Sorting Population values according to 2022 in descending order
total_sorted = total_df.sort_values(by='2022', ascending=False)
```

```python
top_popl = total_sorted.head(20)
```

```python
top_popl.set_index('Country Name', inplace = True)
```

```python
import warnings

# Ignore all warnings
warnings.simplefilter("ignore")
```

### Visualizing the Population Rise in world's most populated areas

```python
top_popl.drop(['World', 'IDA & IBRD total' , 'Low & middle income', 'Middle income'], axis = 0, inplace = True)
```

```python
top_popl.drop(['IBRD only','Early-demographic dividend'], axis = 0, inplace = True)
```

```python
top_popl_transpose = top_popl.transpose()
```

```python
matplotlib.rcParams['font.size'] = 10
plt.figure(figsize=(9,6))
plt.title('Rise in population')
sns.heatmap(top_popl_transpose, cmap='Blues', fmt=".0f", cbar_kws={'label': 'Population'});
plt.ylabel('Years')
plt.xlabel('Countries');
```

Conclusion : The heatmap demonstrates the rise in population for topmost populated areas in the world all along from 1960 to 2022

### Analyzing the total population growth for India

```python
india_df = top_popl[top_popl.index =='India']
india_df
```

```python
world_years = india_df.columns
```

```python
world_population_data = india_df.loc[india_df.index[0]]
```

```python
world_population_data
```

```python
plt.figure(figsize=(12, 8));
plt.plot(world_years, world_population_data, 'o--r');
plt.bar(world_years, world_population_data, color='rosybrown');
plt.title(f'Population Over the Years - {india_df.index[0]}');
plt.xlabel('Year');
plt.ylabel('Population (in Millions )');
plt.xticks(rotation=90);
```

### Comparing Indian male population with Highest male populated country i.e Qatar

```python
qatar_male = male_df[male_df.index  == 'Qatar']
qatar_years = qatar_male.columns
qatar_male_population_data = qatar_male.loc[qatar_male.index[0]]
```

Conclusion - As by this combined line graph we can state that there was drastic increase in the male population in Qatar while India's male population propagated in a straight continuous manner
